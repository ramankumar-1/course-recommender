```python
# All the necessary packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup as bs 
import requests as req
import math
```

```python
seed_url=r"https://archive.nptel.ac.in/noc/noc_courselist.html"
```

```python
# different subject codes present in the URLs of NPTEL courses

subjects={"ae":"Aerospace Engineerig",
        "ag":"Agriculture",
        "ar":"Architecture",
          "bt":"Biotechnology",
          "ch":"Chemical Engineering",
          "cy":"Chemistry & Biology",
          "ce":"Civil Engineering",
          "cs":"Computer Science",
          "ee":"Electrical Engineering",
          "de":"Engineering Design",
          "hs":"Humanities",
          "mg":"Management",
          "ma":"Mathematics",
          "me":"Mechanical Engineering",
          "mm":"Metallurgy & Mining",
          "ge":"Multidiscplinary",
          "oe":"Ocean Engineering",
          "ph":"Physics",
          "te":"Textile Engineering",
          "lw":"Law"}
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
r=req.get(seed_url,headers=headers)
```

```python
# creating BeautifulSoup object

soup=bs(r.text,"html.parser")
```

```python
title=[]
abstract=[]
subject=[]
duration=[] #No. of weeks
course_links=[]
```

```python
# crawling the root page

root="https://archive.nptel.ac.in" #as the course links are Relative URLs

course_rows=soup.find_all("tr")

for i in range(2,len(course_rows)):
  try:
    course_name=course_rows[i].td.text
    if course_name in title: #avoiding crawling the same courses
      continue
    else:
      course_link=course_rows[i].a["href"]
      if course_link.startswith("/noc/courses/"):
        
        course_info_url=root+course_link

        soup2=bs(req.get(course_info_url,headers=headers).text,"html.parser")

        # Reading the abstract of the course
        abstract.append(soup2.find_all("div","c_abstract")[0].p.text)

        # adding the basic information of the course
        course_links.append(course_link)
        title.append(course_name)
        subject.append(course_link[-4:-2])
        duration.append(int(course_rows[i].find_all("td")[3].text.split()[0]))

  except:
    continue
```

```python
len(course_links)
```

```python
# Storing the scraped data in a pandas dataframe

df=pd.DataFrame({"title": title,"abstract": abstract,"subject": subject,"duration": duration,"url":course_links})
df
```

```python
# data preprocessing

# removing stop words
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download("punkt")
nltk.download("stopwords")
```

```python
stop=stopwords.words("english")
print(stop)
```

```python
for i in range(len(df)):
  try:
    df.abstract[i]=[w for w in word_tokenize(df.abstract[i].lower()) if w.isalpha() or w.isnumeric()]
    df.abstract[i]=" ".join(df.abstract[i])
  except:
    continue
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

vec = TfidfVectorizer()

tfidf=vec.fit_transform(df.abstract)
```

```python
# Cosine Similarity for Content-Based filtering

from sklearn.metrics.pairwise import cosine_similarity

cosine=cosine_similarity(tfidf) #cosine similarities of all courses with each other

cosine_df=pd.DataFrame(cosine)

def cosine_sim(course_index):

  # selecting the top 10 similar courses
  index_recommended_courses=cosine_df.iloc[course_index].sort_values(ascending=False).index.tolist()[1:11]

  print("Recommended Courses for",df.title[course_index],"\n")

  recommended_course_names=df.title.iloc[index_recommended_courses]
  print(recommended_course_names)
```

```python
# Recommendation

cosine_sim(935)
```

```python
# Searching courses by keywords

search_kw=str(input("Enter Keywords: ")).lower()

search_kw=" ".join([w for w in search_kw.split() if not (w in stop) and (w.isalpha() or w.isnumeric())])
search_kw
```

```python
kw_tfidf=vec.transform([search_kw])
kw_tfidf
```

```python
# cosine similarity of the courses

kw_cosine_df=pd.DataFrame(cosine_similarity(tfidf, kw_tfidf))
```

```python
index_recommended_courses=kw_cosine_df.iloc[:,0].sort_values(ascending=False).index.tolist()[1:11]
recommended_course_names=df.title.iloc[index_recommended_courses]
print(recommended_course_names)
```
